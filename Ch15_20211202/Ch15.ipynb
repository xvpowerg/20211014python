### 圖表標題與座標軸標籤
+ 使用 plt.title( ) 方法顯示圖表標題。
+ 使用 plt.xlim 設置 x 座標軸範圍。

## Bar 與 Stacked Bar
+ Bar  長條圖：可以垂直或水平繪製
+ Stacked  堆疊長條圖：多項資料的數據一一堆疊，可顯示每一長條的相對組成。

matplotlib.pyplot.bar(x, height, width=0.8, bottom=None, *, align='center', data=None, **kwargs)

https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.bar.html

```python
import matplotlib.pyplot as plt
import numpy as np
city = ['Delhi', 'Beijing', 'Washington', 'Tokyo', 'Moscow']
pos = np.arange(len(city))
Happiness_Index = [60, 40, 70, 65, 85]
plt.bar(pos, Happiness_Index, color='blue', edgecolor='black')
plt.xticks(pos, city)
plt.xlabel('City', fontsize=16)
plt.ylabel('Index', fontsize=16)
plt.title('Barchart ', fontsize=20)
plt.show( )
```

### 基本水平長條圖
matplotlib.pyplot.barh(y, width, height=0.8, left=None, *, align='center', **kwargs)

https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.barh.html

```python
import matplotlib.pyplot as plt
import numpy as np
city = ['Delhi', 'Beijing', 'Washington', 'Tokyo', 'Moscow']
pos = np.arange(len(city))
Happiness_Index = [60, 40, 70, 65, 85]
fig, ax = plt.subplots()# 建立subploat
b = ax.barh(range(len(city)), Happiness_Index) # 建立len(city)組橫方圖
ax.set_yticks(range(len(city)))#一定要給不然圖形會產生錯誤資訊
ax.set_yticklabels(city, fontsize=16)
plt.show()
```

### 堆疊長條圖

```python
import matplotlib.pyplot as plt
import numpy as np
city = ['Delhi', 'Beijing', 'Washington', 'Tokyo', 'Moscow']
Gender = ['Male', 'Female']
pos = np.arange(len(city))
Happiness_Index_Male = [60, 40, 70, 65, 85]
Happiness_Index_Female = [30, 60, 70, 55, 75]
plt.bar(pos, Happiness_Index_Male, color='blue', edgecolor='black')
#Happiness_Index_Female 在 Happiness_Index_Male 底部
plt.bar(pos, Happiness_Index_Female, color='pink', edgecolor='black',bottom=Happiness_Index_Male)
plt.xticks(pos, city)
plt.xlabel('City', fontsize=16)
plt.ylabel('Happiness_Index', fontsize=16)
plt.title('Stacked Barchart ', fontsize=18)
plt.legend(Gender, loc=2)
plt.show( )
```

### 分組長條圖

```python
import matplotlib.pyplot as plt
import numpy as np
city = ['Delhi', 'Beijing', 'Washington', 'Tokyo', 'Moscow']
Gender = ['Male', 'Female']
pos = np.arange(len(city))
bar_width = 0.4
Index_Male = [60, 40, 70, 65, 85]
Index_Female = [30, 60, 70, 55, 75]
plt.bar(pos, Index_Male, bar_width, color='blue', edgecolor='black')
# pos + bar_width設定pink的bar在blue的右邊 
plt.bar(pos + bar_width, Index_Female, bar_width, color='pink', edgecolor='black') 
plt.xticks(pos, city)
plt.xlabel('City', fontsize=16)
plt.ylabel('Happiness_Index', fontsize=16)
plt.title('Stacked Barchart ', fontsize=18)
plt.legend(Gender, loc=2)
plt.show( )
```

### Scatter 與 Hist
+ 散佈圖 Scatter：
  + 有助於揭露兩種變數資料的關係  例如國民平均壽命與生育率。
  + 不像折線圖 各點之間不會直線連接。

+ Hist 直方圖：顯示資料頻率或著數值資料各類別範圍中的資料。

[Scatter](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.scatter.html?highlight=scatter#matplotlib.pyplot.scatter)
[Hist](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.hist.html?highlight=hist#matplotlib.pyplot.hist)

```python
import matplotlib.pyplot as plt
import numpy as np
h=np.array([159,169,174,165,163,155,180,190])
w=np.array([55,45,70,60,65,45,72,50])
print("身高的平均:",np.mean(h))
print("身高的中位:",np.median(h))
print("體重的平均:",np.mean(w))
print("體重的中位:",np.median(w))
#'散佈圖
pos = np.arange(len(h))
plt.scatter(h,w)
plt.show( )
```

### Pie 與 Box chart
+ Pie 圓餅圖：顯示變數的分布，例如各區域的銷售量，每一切片大小式依照每個變數的相對強度呈現。
+ Box chart 盒鬚圖：顯示變數分布，方塊顯示中半段數值，兩側的鬚線向兩端延伸：
  + 將多個資料傳遞給某矩陣資料，再使用 boxplot 函數繪製。
  + boxplot( ) 函數的第一個參數為矩陣資料，第二個參數 patch_artist = True 代表填充，第三個參數獲取要繪製的標籤。
  + boxplot( ) 函數接受參 s 數  vert = 0，它繪製水平框圖。

### Pie
+ 派圖  Pie Plots 使用一個完整圓形顯示統計資料。
  + 可增加 explode 製作突增值進行突出來強調顯示
  + 回傳資料中的 patches 代表傳回的色塊物件。
  + 回傳資料中的 texts 代表傳回的各標籤文字的座標與字串。
  + 使用 legend( ) 顯示圖例，第一個參數是色塊物件，第二個是標籤文字，loc 代表位置，best 代表最佳顯示位置。
  + counterclock 是否為逆時針

```python
import matplotlib.pyplot as plt
values = [60, 80, 90, 55, 10, 30]
colors = ['b', 'g', 'r', 'c', 'm', 'y']
labels = ['US', 'UK', 'India', 'Germany', 'Australia', 'South Korea']
explode = (0.1, 0, 0, 0, 0, 0)
plt.pie(values, colors=colors, labels=values, explode=explode, counterclock=False, shadow=True)
plt.title('Population Density Index')
plt.legend(labels, loc=3)
plt.show()
```

## boxplot(盒鬚圖、盒式圖、盒狀圖或箱線圖)

它能顯示出一組數據的最大值、最小值、中位數、及上下四分位數。

```python
import matplotlib.pyplot as plt
import numpy as np
value1 = [82, 76, 24, 40, 67, 62, 75, 78, 71, 32, 98, 89, 78, 67, 72, 82, 87, 66, 56, 52]
value2 = [62, 5, 91, 25, 36, 32, 96, 95, 3, 90, 95, 32, 27, 55, 100, 15, 71, 11, 37, 21]
value3 = [23, 89, 12, 78, 72, 89, 25, 69, 68, 86, 19, 49, 15, 16, 16, 75, 65, 31, 25, 52]
value4 = [59, 73, 70, 16, 81, 61, 88, 98, 10, 87, 29, 72, 16, 23, 72, 88, 78, 99, 75, 30]
data = np.array(value2)
print(np.median(data))
print(data.max())
print(data.min())
box_plot_data = [value1, value2, value3, value4]
plt.boxplot(box_plot_data)
plt.show( )
## 黃線為中位數
```

## Series
類似於一個dictionary 結構包含key與value 

Series的key對應到 index 而value對應到data

```python
import pandas as pd
import numpy as np
data = [10,11,12]
s1 = pd.Series(data = data,index=["A","B","C"],dtype=np.uint16)
data2 = {"A":92,"B":83,"C":75}
s2 = pd.Series(data2)
print(s1)
print(s2)
```
